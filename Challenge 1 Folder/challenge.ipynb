## Tokenization

```python
spacy.cli.download("xx_ent_wiki_sm")
```

```python
nlp_model = spacy.load("xx_ent_wiki_sm")
```

```python
nlp_model._path
```

```python
def tokenize_text(text):
   doc = nlp_model(text)
   return [token.text for token in doc]
```

```python
train_dataset['banglish_tokens'] = train_dataset['rm'].apply(tokenize_text)

train_dataset
```

```python
train_dataset['bangla_tokens'] = train_dataset['bn'].apply(tokenize_text)

train_dataset
```

```python
val_dataset['banglish_tokens'] = val_dataset['rm'].apply(tokenize_text)
val_dataset['bangla_tokens'] = val_dataset['bn'].apply(tokenize_text)
```

```python
val_dataset
```

## Cleaing

```python
def clean_dataset(dataset, min_length=5, max_length=50):
    cleaned_dataset = dataset[(dataset['rm'].str.split().str.len() >= min_length) &
                              (dataset['rm'].str.split().str.len() <= max_length)]
    return cleaned_dataset
```

```python
train_dataset = clean_dataset(train_dataset)

train_dataset
```

```python
train_dataset.isnull().sum()
```

```python
val_dataset = clean_dataset(val_dataset)

val_dataset
```

# Training

```python
from transformers import MBartForConditionalGeneration, MBartTokenizer
```

```python
from datasets import Dataset

train_dataset = Dataset.from_pandas(train_dataset)
val_dataset = Dataset.from_pandas(val_dataset)
```

```python
model_name = "facebook/mbart-large-50-many-to-many-mmt"
```

```python
tokenizer = MBartTokenizer.from_pretrained(model_name)
```

```python
model = MBartForConditionalGeneration.from_pretrained(model_name)
```

```python
from transformers import  Seq2SeqTrainer, Seq2SeqTrainingArguments
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir='./results',
    eval_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=3,
    predict_with_generate=True
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    processing_class=tokenizer,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)
```

```python
trainer.train()
```

```python
from transformers import MBartForConditionalGeneration, MBartTokenizer

model = MBartForConditionalGeneration.from_pretrained('./banglish_to_bangla_model')
tokenizer = MBartTokenizer.from_pretrained('./banglish_to_bangla_tokenizer')
```

```python
def translate_banglish_to_bangla(text):
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True)

    translated_tokens = model.generate(**inputs)

    translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    return translated_text
```

```python
banglish_text = "Ami Banglay gan gai"
bangla_translation = translate_banglish_to_bangla(banglish_text)
print("Translated Bangla:", bangla_translation)
```
